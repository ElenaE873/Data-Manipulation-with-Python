### Mergining of Datasets

#### Inner Join

- matches of rows/data from both dataframes by the unique column

```python
#mergining two dataframe on column
df_merged = df_1.merge(df_2, on='column')

#it also works this way
df_merged = df_1.merge(df_2, on='column', how='inner')
```

- here df_1 is considered as a "left_table"(similar to sql) where the data in the first df_1 dataframe will appear first, with the columns from the second dataframe df_2 that matched to the right side of the first original table columns

#### One-to-One relationship

- every row in the left dataframe (table) is related to only one row in the right dataframe (table)

#### One-to-Many Relationships

- every rows in the left dataframe (table) is related to **1 or more** rows in the right dataframe (table)
- in this scenario the numbe of rows will certainly increase, and in some cases, greatly increase

```python
#setting the suffixes of the left/right (tables) dataframes for overlapping columns
#this allows the ability to spot the one-to-many records values by having the suffixes identified
df_merged_onetomany = df_1.merge(df_2, on='column', suffixes=('_df_1', '_df_2'))
```

#### Multiple Inner Joins
