#### Left Join

- the result will show the original left dataframe (table) records along with any matches with he right dataframe record columns

```python
#merge with a left join
df_merge_left = df_1.merge(df_2, on='column', how='left')
```

#### Right Join

- the right join is basicly a flip of the left join
- **but there are some distinct differences in the code** as compared to the inner join and left joins WHEN the column names are not the same name

```python
#merge with a right join
df_merge_right = df_1.merge(df_2, how='right',
                           left_on='column_df1', #this is the column name on left table
                            right_on="column_df2") #this is the column name on the right table
```

#### Full Outer Join

```python
#merge with an outer join
df_merge_fullouter = df_1.merge(df_2, on='column', how='outer', suffixes=('_df1', '_df2'))
```
