#### Wide to Long Transformation

Arguments of the .melt() method
- id_vars: identifier variables columns
- the rest of the column are "melted"
    - one column containing the variable names
    - and the other column containing the values for each variable pertaining to identifier columns

##### when not all columns will be needed to be "melted"

Additional arguments:
- value_vars:  takes the names of the columns that are of interested to be melted
- var_name:  the name to be used for the variable column
- value_name: the name of the values column

```python
#code example
df.melt(id_vars='column1', value=vars=['column2','column3'], 
        var_name='nameofcol', value_name='nameofcol')
```
