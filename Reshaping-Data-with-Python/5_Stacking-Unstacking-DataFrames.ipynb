#### Handling Missing Data / Values with Unstacking/Stacking

- Missing values occurs in stacking when values are not found as stacking/unstacking occurs

##### unstacking

```python
#filling in missing values during unstacking
df.unstack(level='column', fill_value="None")
```

##### stacking

- occurs when combinations of index/column values are missing

```python
#in order to keep the missing values when stackign occurs, need to set dropna to False
df.stack(dropna=False)

#and then fill them in with a value as deemed appropriate
df.stack(dropna=False).fillna(0)
```
